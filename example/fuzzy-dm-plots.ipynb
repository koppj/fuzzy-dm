```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load GLoBES output file for CPV sensitivity
delta, chi2_noDM, chi2_scalar, chi2_vector_pol, chi2_vector_unpol = np.loadtxt("dm-test-cpv.dat", comments="#", unpack=True)
chi2_min_noDM         = np.min(chi2_noDM)
chi2_min_scalar       = np.min(chi2_scalar)
chi2_min_vector_pol   = np.min(chi2_vector_pol)
chi2_min_vector_unpol = np.min(chi2_vector_unpol)

# make plot
plt.plot(delta, np.sqrt(chi2_noDM - chi2_min_noDM),                 ls=':', label="no DM")
plt.plot(delta, np.sqrt(chi2_scalar - chi2_min_scalar),             ls='-', label="scalar DM")
plt.plot(delta, np.sqrt(chi2_vector_pol - chi2_min_vector_pol),     ls='-', label="vector DM (polarized)")
#plt.plot(delta, np.sqrt(chi2_vector_unpol - chi2_min_vector_unpol), ls='-', label="vector DM (unpolarized)")
plt.xlabel("Dirac CP phase $\delta_{CP}$")
plt.ylabel("$\sigma = \sqrt{\chi^2}$")
plt.legend()
plt.show()
```

```python
# load GLoBES output file for DM sensitivity
mchi = 1e-22
y, chi2_scalar, chi2_vector_pol, chi2_vector_unpol = np.loadtxt("dm-test-dmchi2.dat", comments="#", unpack=True)

# make plot
plt.figure(figsize=(7,5))
plt.plot(y/mchi, np.abs(chi2_scalar),       ls='-', label="scalar DM")
plt.plot(y/mchi, np.abs(chi2_vector_pol),   ls='-', label="vector DM (polarized)")
plt.plot(y/mchi, np.abs(chi2_vector_unpol), ls='-', label="vector DM (unpolarized)")
plt.axhline(1, color="gray", ls=":")
plt.axhline(4, color="gray", ls=":")
plt.axhline(9, color="gray", ls=":")
plt.ylim(0,30)
plt.xscale("log")
plt.xlabel("$y / m_{DM}$ [eV$^{-1}$]")
#plt.ylabel("$\sigma = \sqrt{\chi^2}$")
plt.ylabel("$\Delta\chi^2$")
plt.annotate("1$\sigma$", xy=(3e-14,1.2), color="gray")
plt.annotate("2$\sigma$", xy=(3e-14,4.2), color="gray")
plt.annotate("3$\sigma$", xy=(3e-14,9.2), color="gray")
plt.legend(loc="upper center")
plt.grid(axis="x")
plt.show()
```
