```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load GLoBES output file for CPV sensitivity
delta, chi2_noDM, chi2_scalar, chi2_vector_pol, chi2_vector_unpol = np.loadtxt("dm-test-cpv.dat", comments="#", unpack=True)
chi2_min_noDM         = np.min(chi2_noDM)
chi2_min_scalar       = np.min(chi2_scalar)
chi2_min_vector_pol   = np.min(chi2_vector_pol)
chi2_min_vector_unpol = np.min(chi2_vector_unpol)

# make plot
plt.plot(delta, np.sqrt(chi2_noDM - chi2_min_noDM),                 ls=':', label="no DM")
plt.plot(delta, np.sqrt(chi2_scalar - chi2_min_scalar),             ls='-', label="scalar DM")
plt.plot(delta, np.sqrt(chi2_vector_pol - chi2_min_vector_pol),     ls='-', label="vector DM (polarized)")
#plt.plot(delta, np.sqrt(chi2_vector_unpol - chi2_min_vector_unpol), ls='-', label="vector DM (unpolarized)")
plt.xlabel("Dirac CP phase $\delta_{CP}$")
plt.ylabel("$\sigma = \sqrt{\chi^2}$")
plt.legend()
plt.show()
```

```python
# load GLoBES output file for DM sensitivity
delta, chi2_scalar, chi2_vector_pol, chi2_vector_unpol = np.loadtxt("dm-test-dmchi2.dat", comments="#", unpack=True)

# make plot
plt.plot(delta, np.sqrt(np.abs(chi2_scalar))      , ls='-', label="scalar DM")
plt.plot(delta, np.sqrt(np.abs(chi2_vector_pol)),   ls='-', label="vector DM (polarized)")
plt.plot(delta, np.sqrt(np.abs(chi2_vector_unpol)), ls='-', label="vector DM (unpolarized)")
plt.xscale("log")
plt.xlabel("$y / m_\text{DM}$")
plt.ylabel("$\sigma = \sqrt{\chi^2}$")
plt.ylim(0,10)
plt.show()
```
